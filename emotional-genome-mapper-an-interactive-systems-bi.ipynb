```python
# ==============================
# EMOTIONAL GENOME MAPPER
# Complete Interactive Version
# ==============================

# Install required libraries
!pip -q install ipywidgets seaborn

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import ipywidgets as widgets
from IPython.display import display, clear_output
import networkx as nx

plt.style.use("dark_background")

# ------------------------------
# Biological Simulation Engine
# ------------------------------

def simulate_genome(stress, sleep, diet, exercise):

    # Baseline gene expression
    baseline = {
        "TP53 (Tumor Suppressor)": 100,
        "BDNF (Mood Regulator)": 100,
        "IL6 (Inflammation Marker)": 100
    }

    # Biological relationships (simulated)
    cortisol = stress * 0.8 - sleep * 0.3
    cortisol = max(cortisol, 0)

    methylation = cortisol * 0.05

    gene_expression = {
        gene: max(baseline[gene] - methylation + diet*0.2 + exercise*0.3, 0)
        for gene in baseline
    }

    inflammation = gene_expression["IL6 (Inflammation Marker)"] * 1.2
    risk_index = (stress*0.5 + inflammation*0.02) - (exercise*0.3)

    return gene_expression, inflammation, risk_index


# ------------------------------
# Visualization Function
# ------------------------------

def run_simulation(stress, sleep, diet, exercise):

    clear_output(wait=True)

    gene_expression, inflammation, risk_index = simulate_genome(
        stress, sleep, diet, exercise
    )

    print("\n🧠 Emotional Genome Mapper Simulation\n")

    # Heatmap
    df = pd.DataFrame(
        gene_expression.values(),
        index=gene_expression.keys(),
        columns=["Expression Level"]
    )

    plt.figure(figsize=(6,4))
    sns.heatmap(df, annot=True, cmap="coolwarm", linewidths=1)
    plt.title("Gene Expression Heatmap")
    plt.show()

    # Risk Graph
    plt.figure(figsize=(6,4))
    plt.bar(["Inflammation Score", "Cancer Risk Index"],
            [inflammation, risk_index])
    plt.title("Biological Risk Indicators")
    plt.show()

    # Network Visualization
    G = nx.DiGraph()
    G.add_edges_from([
        ("Stress", "Cortisol"),
        ("Cortisol", "DNA Methylation"),
        ("DNA Methylation", "Gene Expression"),
        ("Gene Expression", "Inflammation"),
        ("Inflammation", "Disease Risk")
    ])

    plt.figure(figsize=(6,6))
    pos = nx.spring_layout(G)
    nx.draw(G, pos, with_labels=True, node_size=3000,
            node_color="purple", font_size=10)
    plt.title("Biological Cascade Network")
    plt.show()

    print(f"\n🔥 Inflammation Score: {round(inflammation,2)}")
    print(f"⚠ Cancer Risk Index: {round(risk_index,2)}")


# ------------------------------
# Beautiful Interactive UI
# ------------------------------

stress_slider = widgets.IntSlider(
    value=30, min=0, max=100, step=1,
    description="Stress Level",
    style={'description_width': 'initial'}
)

sleep_slider = widgets.IntSlider(
    value=7, min=0, max=12, step=1,
    description="Sleep Hours",
    style={'description_width': 'initial'}
)

diet_slider = widgets.IntSlider(
    value=50, min=0, max=100, step=1,
    description="Diet Quality",
    style={'description_width': 'initial'}
)

exercise_slider = widgets.IntSlider(
    value=30, min=0, max=100, step=1,
    description="Exercise Level",
    style={'description_width': 'initial'}
)

button = widgets.Button(description="Run Simulation",
                        button_style='success')

def on_button_clicked(b):
    run_simulation(
        stress_slider.value,
        sleep_slider.value,
        diet_slider.value,
        exercise_slider.value
    )

button.on_click(on_button_clicked)

display(
    widgets.VBox([
        widgets.HTML("<h2 style='color:cyan;'>🧬 Emotional Genome Mapper</h2>"),
        stress_slider,
        sleep_slider,
        diet_slider,
        exercise_slider,
        button
    ])
)
```
